Q2.

**LIST**
- List is mutable.

- Created using square brackets **[ ]**.

- Faster modification is possible.

- Methods like append() and remove() are available.

**TUPLE**
- Tuples are immutable (cannot be modified).

- Created using parentheses **( )**.

- More memory-efficient and faster for fixed data.

- Limited methods such as count() and index().

- Suitable for data that must remain constant, such as coordinates, configuration settings, or dictionary keys.
